## Actividad 1: Abrir una página web y extraer el título

Objetivo: Aprender a abrir una página y extraer el título.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Usa el Service para gestionar el chromedriver
service = Service(ChromeDriverManager().install())

# Inicializa el navegador
driver = webdriver.Chrome(service=service)
```

```python
# Abre una página web
driver.get("https://quotes.toscrape.com/")

# Obtén el título de la página
title = driver.title
print(f"El título de la página es: {title}")

# Cierra el navegador
driver.quit()
```
