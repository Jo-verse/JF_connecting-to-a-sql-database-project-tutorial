## Actividad 3: Realizar una búsqueda en eBay y extraer los primeros 10 resultados

Objetivo: Simular una búsqueda en eBay y extraer información de los primeros 10 resultados de productos.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Usa el Service para gestionar el chromedriver
service = Service(ChromeDriverManager().install())

# Inicializa el navegador
driver = webdriver.Chrome(service=service)

# Abre eBay
driver.get("https://www.ebay.com/")

# Encuentra la barra de búsqueda y escribe el término de búsqueda
search_box = driver.find_element(By.ID, "gh-ac")
search_box.send_keys("laptop")
search_box.send_keys(Keys.RETURN)

# Espera para asegurar que los resultados se carguen
time.sleep(3)

# Encuentra los primeros 10 resultados de productos
items = driver.find_elements(By.CSS_SELECTOR, ".s-item")[:10]

# Extrae el título, el precio y el enlace de cada producto
for item in items:
    title = item.find_element(By.CSS_SELECTOR, ".s-item__title").text
    price = item.find_element(By.CSS_SELECTOR, ".s-item__price").text
    link = item.find_element(By.CSS_SELECTOR, ".s-item__link").get_attribute("href")
    
    print(f"Producto: {title}")
    print(f"Precio: {price}")
    print(f"Enlace: {link}")
    print("-" * 40)

# Cierra el navegador
driver.quit()
```

## Actividad 4: Manejo de Páginas con Carga Dinámica (Scroll Automático)

Objetivo: Realizar scroll para cargar contenido dinámicamente y extraer información.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Usa el Service para gestionar el chromedriver
service = Service(ChromeDriverManager().install())

# Inicializa el navegador
driver = webdriver.Chrome(service=service)

# Accede a la página con scroll infinito
driver.get("https://scrapingclub.com/exercise/list_infinite_scroll/")

#obtener la altura de la página
last_height = driver.execute_script("return document.body.scrollHeight")

# inciamos un buble para hacer scroll hasta el final de la página
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Buscamos todos los elementos con la clase item
items = driver.find_elements(By.CSS_SELECTOR, ".item")

# Imprimimos el texto de cada elemento
for item in items:
    print(item.text)

# Cierra el navegador
driver.quit()
```
